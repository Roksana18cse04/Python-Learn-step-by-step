```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# to define independent and dependent variable
t= np.arange(0,1,0.001)
t
```

```python
xt=np.sin(2*np.pi*3*t)
xt
```

```python
# to plot the signal
plt.plot(t,xt)
plt.title('SINE WAVE')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

# COS WAVE VIEW

```python
xt=np.cos(2*np.pi*3*t)
xt
```

```python
# to plot the COS signal
plt.plot(t,xt)
plt.title('SINE WAVE')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```
