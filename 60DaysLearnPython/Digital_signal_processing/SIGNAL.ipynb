# To Plot Sampled sine and cos signal

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# to define independent and dependent variable
n =np.arange(0,1,0.01)
xn =np.sin(2*np.pi*3*n)
xn
```

```python
# to plot the signal
plt.plot(n,xn)
plt.title('SINE WAVE')
plt.xlabel('Time')
plt.ylabel('Amplitude')
```

```python
plt.stem(n,xn)
plt.title('sampled i sine wave')
plt.title('SINE WAVE')
plt.xlabel('------------>n')
plt.ylabel('Amplitude')
```
